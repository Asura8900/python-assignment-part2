***FOR TASK 2***

*Requirement 1*: To write logic to add an item to the cart

```python
#  Creating the cart as a list of dictionaries, starting empty
cart = []

# Options to decide what action to take
print("Type '1' to add an item")
print("Type '2' to remove an item")
print("Type '3' to update quantity of an item")
print("Type '4' to checkout")

while True:    
    action_choice = int(input("Enter choice 1-4" ))
    print("========================================================================================================================")
    if action_choice == 1: # For requirement 1
        item_name = input("Enter item name to add").title()
        if item_name not in menu:
            print(f"{item_name} is not in menu")
        elif menu[item_name]["available"] == False:
            print(f"{item_name} is currently unavailable")
        else:
            quantity = int(input("Enter quantity"))
            for cart_item in cart:
                if cart_item["item"] == item_name:
                    cart_item["quantity"] = cart_item["quantity"] + quantity
                    break
            else:
                    cart.append({"item": item_name, "quantity": quantity, "price": menu[item_name]["price"]})
    

    elif action_choice == 2: # For requirement 2
        item_name = input("Enter item name to remove").title()
        for cart_item in cart:
            if cart_item["item"] == item_name:
                cart.remove(cart_item)
                break
        else:
            print(f'{item_name} is not in the cart')
        
    elif action_choice == 3: # For requirement 3
        item_name = input("Enter item name to update").title()
        new_quantity = int(input("Enter updated quantity"))
        for cart_item in cart:
            if cart_item["item"] == item_name:
                cart_item["quantity"] = new_quantity
                break
        else:
            print(f'{item_name} is not in the cart')

    elif action_choice == 4: # For requirement 5
        if len(cart) == 0:
            print("Cart is empty. Add some items before checking out")
        else:
            subtotal = 0
            for cart_item in cart:
                subtotal = subtotal + (cart_item["price"] * cart_item["quantity"])
            
            GST = (5/100) * subtotal
            Total = subtotal + GST
            
            print("========== Order Summary ==========")
            for cart_item in cart:
                cart_item_total = cart_item["quantity"] * cart_item["price"]
                print(f'{cart_item["item"]:20} x{cart_item["quantity"]}    ₹{cart_item_total}')
            print("-----------------------------------")
            print(f'Subtotal     : ₹{subtotal:.2f}')
            print(f'GST (5%)     : ₹{GST:.2f}')
            print(f'Total Payable: ₹{Total:.2f}')
            print("====================================")
            break
    
    print(cart)
    print("========================================================================================================================")
                
```

```python
print("========== revenue per day ==========")
print("      DATE      |   TOTAL REVENUE   ")
highest_total_revenue = 0
for item, value in sales_log.items():
    date = item
    total_revenue = 0
    
    for order in value:
        total_revenue += order["total"]
    print(f'   {date:13}|{total_revenue:10}')

    if highest_total_revenue < total_revenue:
        highest_total_revenue = total_revenue
        highest_total_revenue_date = date


print("=====================================")
print()
print("============================================================")
print(f'The best selling date is {highest_total_revenue_date} with revenue of {highest_total_revenue}')
print("============================================================")
```

```python
item_count_list = {}
for day in sales_log.values():
    for order in day:
        for item in order["items"]:
            if item in item_count_list:
                item_count_list[item] +=1
            else:
                item_count_list[item] = 1

most_ordered = max(item_count_list , key=item_count_list.get)
print(most_ordered)
```
